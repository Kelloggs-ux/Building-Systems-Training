```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
```

# Assignment 8

## Parameters

```
Heating_Loads = ["Opaque Surfaces","Fenestration","Infiltration","Distribution Losses"]
Cooling_Loads = ["Opaque Surfaces","Fenestration","Internal Gains","Infiltration","Distribution Losses"]

Heating_sensible_values = [2850.9,4388.85,2472,2816.4]
Cooling_values = [1330.9,3709.7,576,432.9,4295.2,491]
Cooling_latent_values = [83.6,407.4]

colours = ["#FFFF00","#0000FF","#FF00FF","#00FF00","#00FFFF","#FF0000"]
colours2 = ["#FF6347","#CD5C5C"]
```

## Pie Charts

### Part A

#### Heating

```
figure = plt.figure(figsize=(15, 5))
plt.title("Heating Loads",fontsize=15)
plt.suptitle("(sensible)")
plt.pie(Heating_sensible_values,labels=Heating_Loads,colors=colours,startangle=90,shadow=True,autopct="%1.1f%%")
PieChart_Heating = figure
```

#### Cooling

```
# make figure and assign axis objects
fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
fig.subplots_adjust(wspace=0)

# Total Heat pie chart parameters
Total = sum(Cooling_values)
ratios = [x/Total for x in Cooling_values]
labels = ["Opaque Surfaces","Fenestration","Internal Gains","Infiltration","Distribution Losses","Latent Heat"]
explode = [0, 0, 0, 0, 0, 0.2]
# rotate so that first wedge is split by the x-axis
angle = 8.16
ax1.pie(ratios, autopct='%1.1f%%', startangle=angle,labels=labels, explode=explode,colors=colours)

# Latent Heat pie chart parameters
Total2 = sum(Cooling_latent_values)
ratios_ax2 = [x/Total2 for x in Cooling_latent_values]
labels2 = ["Internal Gains","Air Flow"]
width = .15

ax2.pie(ratios_ax2, autopct='%1.1f%%', startangle=135,labels=labels2, radius=0.5, textprops={'size': 'smaller'},colors=colours2)


ax1.set_title('Total Cooling Loads')
ax2.set_title('Latent Loads')

# use ConnectionPatch to draw lines between the two plots
# get the wedge data
n=5
theta1, theta2 = ax1.patches[n].theta1, ax1.patches[n].theta2
center, r = ax1.patches[n].center, ax1.patches[n].r

# draw top connecting line
x = r * np.cos(np.pi / 180 * theta2) + center[0]
y = np.sin(np.pi / 180 * theta2) + center[1]
con = ConnectionPatch(xyA=(- width / 2, .5), xyB=(x, y),coordsA="data", coordsB="data", axesA=ax2, axesB=ax1)
con.set_color([0, 0, 0])
con.set_linewidth(1.5)
ax2.add_artist(con)

# draw bottom connecting line
x = r * np.cos(np.pi / 180 * theta1) + center[0]
y = np.sin(np.pi / 180 * theta1) + center[1]
con = ConnectionPatch(xyA=(- width / 2, -.5), xyB=(x, y), coordsA="data",coordsB="data", axesA=ax2, axesB=ax1)
con.set_color([0, 0, 0])
ax2.add_artist(con)
con.set_linewidth(1.5)
PieChart_Cooling = fig
```
