# Assignment 1

```
height= 0.8 # m
width= 1.5 # m
Tinf1= 20 #°C
Tinf2= -10 #°C
h1= 10 #W/°C*m^2
h2= 40 #W/°C*m^2

L_SingleGlass= 0.008 # m
L_DoubleGlass= 0.004 # m
L_air= 0.01 # m
k_glass= 0.78 # W/°C*m
k_air= 0.026 # W/°C*m
```

## Question A

```
A=height*width
R_conv1=1/(h1*A)
R_conv2=1/(h2*A)
R_cond=L_SingleGlass/(k_glass*A)

R_Tot=R_conv1+R_cond+R_conv2
Q=(Tinf1-Tinf2)/R_Tot
T1=Tinf1-Q/(h1*A)

print("The Total Thermal Resistance of the windows is determined to be "+str(R_Tot)+" °C/W, resulting in a Heat Transfer across the windows of "+str(Q)+" W. With this, the temperature of the internal face fo the glass si determined to be "+str(T1)+" °C")
```

## Question B

```
A=height*width
R_conv1=1/(h1*A)
R_conv2=1/(h2*A)
R_condAir=L_air/(k_air*A)
R_condG1=L_DoubleGlass/(k_glass*A)
R_condG2=L_DoubleGlass/(k_glass*A)

R_Tot=R_conv1+R_condG1+R_condAir+R_condG2+R_conv2
Q=(Tinf1-Tinf2)/R_Tot
T1=Tinf1-Q/(h1*A)

print("The Total Thermal Resistance of the windows is determined to be "+str(R_Tot)+" °C/W, resulting in a Heat Transfer across the windows of "+str(Q)+" W. With this, the temperature of the internal face fo the glass si determined to be "+str(T1)+" °C")
```
